```python
import numpy as np
import pandas as pd
import calendar
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('https://data.humdata.org/hxlproxy/api/data-preview.csv?url=https%3A%2F%2Fraw.githubusercontent.com%2FCSSEGISandData%2FCOVID-19%2Fmaster%2Fcsse_covid_19_data%2Fcsse_covid_19_time_series%2Ftime_series_19-covid-Confirmed.csv&filename=time_series_2019-ncov-Confirmed.csv')
```

```python
df
```

```python
df = df.drop(['Province/State', 'Lat', 'Long'], axis=1)
```

```python
df = df.sort_values(by = 'Country/Region')
```

```python
df
```

```python
df = df.groupby(['Country/Region']).sum()
```

```python
df
```

```python
df = df.loc[:, df.isin(['0']).mean() <= .7]
```

```python
df
```

```python
plot1 = df.plot(kind='bar', y='3/22/20')
```

```python
plot1.get_figure().savefig('all_country_data.png', dpi = 300)
```

```python
def top_countries_for_date(date, count_countries):
    top_countries = df.nlargest(count_countries, date)
    plot2 = top_countries.plot(y=date, kind='bar')
    plot2.get_figure().savefig('all_country_data.png', dpi = 300)
```

```python
top_countries_for_date('3/20/20', 20)
```

```python
def datewise_for_country(country):
    plot3 = df.loc[country, : ].T.plot(kind = 'line')
    plot3.set(xlabel = country, ylabel = 'Number of Case')
    plot3.get_figure().savefig(country+'.png', dpi = 300)
```

```python
datewise_for_country('Italy')
```

```python
datewise_for_country('India')
```

```python
def continent_plot():
    merge_df = pd.merge(left=df, right=continentDf, left_on='Country/Region', right_on='Country')
    merge_Continent = merge_df.groupby(['Continent'],as_index=False).sum()
    plot4 = merge_Continent.plot(x = 'Continent', y = '3/22/20',kind='pie', labels = merge_Continent['Continent'], legend = False)
    plot4.get_figure().savefig('continent.png', dpi = 300)
```

```python
continent_plot()
```
